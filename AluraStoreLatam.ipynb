### Importación de datos

#1. Análisis de facturación



###Ingreso total por cada tienda

####En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
sum_tienda = tienda["Precio"].sum()
print(sum_tienda)
```

```python
sum_tienda2 = tienda2["Precio"].sum()
print(sum_tienda2)
```

```python
sum_tienda3 = tienda3["Precio"].sum()
print(sum_tienda3)
```

```python
sum_tienda4 = tienda4["Precio"].sum()
print(sum_tienda4)
```

# 2. Ventas por categoría

###En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
venta_cat = tienda.groupby("Categoría del Producto")
suma_cat = venta_cat["Categoría del Producto"].count()
print(suma_cat)
```

```python
venta_cat = tienda2.groupby("Categoría del Producto")
suma_cat = venta_cat["Categoría del Producto"].count()
print(suma_cat)
```

```python
venta_cat = tienda3.groupby("Categoría del Producto")
suma_cat = venta_cat["Categoría del Producto"].count()
print(suma_cat)
```

```python
venta_cat = tienda4.groupby("Categoría del Producto")
suma_cat = venta_cat["Categoría del Producto"].count()
print(suma_cat)
```

# 3. Calificación promedio de la tienda


###En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
califica_tienda1 = tienda["Calificación"].mean().round(2)
print(califica_tienda1)
```

```python
califica_tienda2 = tienda2["Calificación"].mean().round(2)
print(califica_tienda2)
```

```python
califica_tienda3 = tienda3["Calificación"].mean().round(2)
print(califica_tienda3)
```

```python
califica_tienda4 = tienda4["Calificación"].mean().round(2)
print(califica_tienda4)
```

# 4. Productos más y menos vendidos

###En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
menos_vendidos = tienda["Categoría del Producto"].min()
print(f"Los productos menos vendidos son: {menos_vendidos}")

mas_vendidos = tienda["Categoría del Producto"].max()
print(f"Los productos más vendidos son: {mas_vendidos}")
```

```python
menos_vendidos3 = tienda3["Categoría del Producto"].min()
print(f"Los productos menos vendidos son: {menos_vendidos3}")

mas_vendidos3 = tienda3["Categoría del Producto"].max()
print(f"Los productos más vendidos son: {mas_vendidos3}")
```

```python
menos_vendidos4 = tienda4["Categoría del Producto"].min()
print(f"Los productos menos vendidos son: {menos_vendidos4}")

mas_vendidos4 = tienda4["Categoría del Producto"].max()
print(f"Los productos más vendidos son: {mas_vendidos4}")
```

# 5. Envío promedio por tienda

###En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
promedio_t1 = tienda["Costo de envío"].mean().round(2)
print(f"El costo de envío promedio de la tienda 1 es: {promedio_t1}")
```

```python
promedio_t2 = tienda2["Costo de envío"].mean().round(2)
print(f"El costo de envío promedio de la tienda 2 es: {promedio_t2}")
```

```python
promedio_t3 = tienda3["Costo de envío"].mean().round(2)
print(f"El costo de envío promedio de la tienda 3 es: {promedio_t3}")
```

```python
promedio_t4 = tienda4["Costo de envío"].mean().round(2)
print(f"El costo de envío promedio de la tienda 4 es: {promedio_t4}")
```
